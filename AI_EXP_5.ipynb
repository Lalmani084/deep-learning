```python
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
df_experience = pd.read_csv("/content/sentiment_analysis_poem.csv")
```

```python
df_experience['Text'][0]
```

```python
TextBlob(df_experience['Text'][10])
```

```python
TextBlob(df_experience['Text'][10]).sentiment
```

```python
df_experience['Polarity']=df_experience['Text'].apply(lambda x:TextBlob(x).polarity)
df_experience
```

```python
df_experience['Subjectivity']=df_experience['Text'].apply(lambda x:TextBlob(x).subjectivity)
df_experience
```

```python
df_experience['Sentiment'] = df_experience['Polarity'].apply(lambda x : "Positive" if x > 0 else "Neutral" if x ==0 else "Negative")
df_experience
```

```python
sentiment_counts = df_experience['Sentiment']. value_counts()
sentiment_counts
```

```python
sentiment_counts.plot(kind='bar')
plt.title('Sentiment Analysis')
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```
